```python
import pandas as pd

# 데이터 로드
data = pd.read_csv('train-metadata.csv')

# target 값의 분포 확인
print(data['target'].value_counts())

malignant_data = data[data['target'] == 1]

print(malignant_data['isic_id'].head(5))

isic_id = 'ISIC_0082829'
target_value = data.loc[data['isic_id'] == isic_id, 'target']

# 결과 출력
if not target_value.empty:
    print(f"이미지 ID '{isic_id}'에 해당하는 target 값: {target_value.iloc[0]}")
else:
    print(f"이미지 ID '{isic_id}'에 해당하는 데이터가 없습니다.")
```

```python
import pandas as pd
import os

# CSV 파일 로드
metadata_path = 'train-metadata.csv'
metadata = pd.read_csv(metadata_path)

# 데이터 확인
print(metadata.head())  # 첫 5개 행 출력
print(metadata.columns)  # 열 이름 출력

# 이미지 파일 경로 설정
image_folder = 'train-image/image'

# 이미지 파일 샘플 출력
sample_images = metadata['isic_id'].head(5)  # 상위 5개 이미지 ID 확인
for image_id in sample_images:
    print(f"{image_id}.jpg is at {os.path.join(image_folder, image_id + '.jpg')}")
```

```python
from PIL import Image
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator

images = []
labels = []



# 이미지와 레이블 로드
for _, row in metadata.iterrows():
    img_path = os.path.join(image_folder, row['isic_id'] + '.jpg')
    if os.path.exists(img_path):
        # 이미지 로드 및 전처리
        image = Image.open(img_path).resize((128, 128))  # 크기 변경
        image = np.array(image, dtype=np.float32) / 255.0  # 정규화
        images.append(image)
        labels.append(row['target'])

# Numpy 배열로 변환
images = np.array(images)
labels = np.array(labels)

# 데이터 확인
print(f"Images shape: {images.shape}")
print(f"Labels shape: {labels.shape}")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(images, labels, test_size=0.2, random_state=42)

print(f"Training data: {X_train.shape}, {y_train.shape}")
print(f"Validation data: {X_val.shape}, {y_val.shape}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import Adam

# 모델 구성
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')  # 악성 확률 출력
])

# 모델 컴파일
model.compile(optimizer=Adam(learning_rate=1e-3),
              loss='binary_crossentropy',
              metrics=['accuracy'])

print(model.summary())  # 모델 요약 출력
```

```python
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_val, y_val))

# 학습 결과 확인
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
loss, accuracy = model.evaluate(X_val, y_val)
print(f"Validation Loss: {loss}")
print(f"Validation Accuracy: {accuracy}")
```

```python
def classify_image(image_path):
    image = Image.open(image_path).resize((128, 128))
    image = np.array(image, dtype=np.float32) / 255.0
    image = np.expand_dims(image, axis=0)  # 배치 차원 추가
    prediction = model.predict(image)[0][0]
    if prediction < 0.5:
        return "Benign (양성)"
    else:
        return "Malignant (악성)"

# 테스트 이미지 분류
test_image_path = 'train-image/image/ISIC_0082829.jpg'  # 테스트 이미지 경로
result = classify_image(test_image_path)
print(f"Prediction for the test image: {result}")
```

```python

 import os
# CSV 파일 로드
metadata = pd.read_csv('train-metadata.csv')

# 이미지 파일 경로 설정
image_folder = 'train-image/image'

# 이미지 경로 추가
data['image_path'] = data['isic_id'].apply(lambda x: os.path.join(image_folder, f"{x}.jpg"))

# 데이터 확인
print(data.head())
print(data['target'].value_counts())
```

```python
from sklearn.model_selection import train_test_split

# 학습용 및 테스트용 데이터 분리
train_data, test_data = train_test_split(data, test_size=0.2, stratify=data['target'], random_state=42)

# 분리된 데이터 확인
print(f"학습 데이터 개수: {len(train_data)}")
print(f"테스트 데이터 개수: {len(test_data)}")
```

```python
# 학습 데이터 분포
print("학습 데이터 분포:")
print(train_data['target'].value_counts())

# 테스트 데이터 분포
print("\n테스트 데이터 분포:")
print(test_data['target'].value_counts())
```

```python
import tensorflow as tf

# 이미지 로드 및 전처리 함수
def preprocess_image(image_path, label):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [224, 224])  # 크기 조정
    image = image / 255.0  # 정규화
    return image, label

# 학습 및 테스트 데이터 준비
train_paths = train_data['image_path'].values
train_labels = train_data['target'].values

test_paths = test_data['image_path'].values
test_labels = test_data['target'].values

train_dataset = tf.data.Dataset.from_tensor_slices((train_paths, train_labels))
train_dataset = train_dataset.map(preprocess_image).batch(32).shuffle(buffer_size=1000)

test_dataset = tf.data.Dataset.from_tensor_slices((test_paths, test_labels))
test_dataset = test_dataset.map(preprocess_image).batch(32)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.metrics import classification_report

# 간단한 CNN 모델 정의
def create_model(input_shape=(224, 224, 3), num_classes=2):
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dense(num_classes, activation='softmax')  # 두 클래스 (양성, 악성)
    ])
    return model

# 모델 컴파일
model = create_model()
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 데이터 전처리: TensorFlow Dataset 준비
# train_dataset과 test_dataset은 전처리된 데이터셋으로 가정
# 이미 데이터셋이 준비되었다면 아래 코드는 건너뛰어도 됩니다.

# 학습 데이터 로드 예시
train_dataset = tf.data.Dataset.from_tensor_slices((train_data['image_path'].values, train_data['target'].values))
test_dataset = tf.data.Dataset.from_tensor_slices((test_data['image_path'].values, test_data['target'].values))

# 이미지 전처리 함수
def preprocess_image(image_path, label):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [224, 224])
    image = image / 255.0  # 정규화
    return image, label

# 데이터셋 전처리 및 배치 구성
train_dataset = train_dataset.map(preprocess_image).batch(32).shuffle(1000)
test_dataset = test_dataset.map(preprocess_image).batch(32)

# 모델 학습
model.fit(train_dataset, epochs=5)

# 모델 평가
test_loss, test_accuracy = model.evaluate(test_dataset)
print(f"테스트 정확도: {test_accuracy:.2f}")

# 테스트 데이터로 예측
y_true = []
y_pred = []

for images, labels in test_dataset:
    preds = model.predict(images)
    y_true.extend(labels.numpy())
    y_pred.extend(tf.argmax(preds, axis=1).numpy())

# 성능 보고서 출력
print(classification_report(y_true, y_pred, target_names=['Benign', 'Malignant']))
```
